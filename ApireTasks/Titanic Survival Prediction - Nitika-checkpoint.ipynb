```python
#Model Training
dataframe
```

```python
X = dataframe.iloc[:,0:6]
y = dataframe.iloc[:,-1]
print('\n-------------------------------- X Information --------------------------------\n')
X.info()
print('\--------------------------------- y Information --------------------------------\n')
y.info()
```

```python
X_train, X_test,y_train, y_test = train_test_split(X,y ,random_state=104, test_size=0.30, shuffle=True)
print(len(X_train),len(y_train))
print(len(X_test),len(y_test))
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=3,p=2,metric='euclidean')
knn_fit = knn.fit(X_train,y_train)
knn_fit
```

```python
y_pred = knn.predict(X_test)
y_pred
```

```python
# Finding Accuracy of Prediction Model

acc = knn.score(X_train,y_train,sample_weight=None)
print("The accuracy is: ",round(acc*100,3),"%")
```
